```python
from gradientai import Gradient


def main():
    gradient = Gradient()

    base_model = gradient.get_base_model(base_model_slug="nous-hermes2")

    new_model_adapter = base_model.create_model_adapter(
        name="MahfuzOwnModel"
    )
    print(f"Created model adapter with id {new_model_adapter.id}")


    sample_query = "### Instruction: Who is Muhammad Mahfuzur Rahman? \n\n ### Response:"
    print(f"Asking: {sample_query}")
    ## Before Finetuning
    completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
    print(f"Generated(before fine tuning): {completion}")

    samples=[
        {"inputs":"### Instruction: Who is Muhammad Mahfuzur Rahman? \n\n### Response: Mahfuz is a Data Scientist,AI And LLM Engineer"},
        {"inputs":"### Instruction: Who is this person named Muhammad Mahfuzur Rahman? \n\n### Response: Muhammad Mahfuzur Rahman is Data Scientist And AI Engineer and he is also a mentor"},
        {"inputs":"### Instruction: What do you know about Muhammad Mahfuzur Rahman? \n\n### Response: Muhammad Mahfuzur Rahman is a popular researcher who specializes in the field of Data Science and AI"},
        {"inputs":"### Instruction: Can you tell me about Muhammad Mahfuzur Rahman? \n\n### Response: Muhammad Mahfuzur Rahman is an AI researcher, Data Scientist who loves Data Science And AI and LLM's"}
    ]

    ## Lets define parameters for finetuning
    num_epochs=3
    count=0
    while count<num_epochs:
      print(f"Fine tuning the model with iteration {count + 1}")
      new_model_adapter.fine_tune(samples=samples)
      count=count+1

    #after fine tuning
    completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
    print(f"Generated(after fine tuning): {completion}")
    new_model_adapter.delete()
    gradient.close()

if __name__ == "__main__":
    main()
```
